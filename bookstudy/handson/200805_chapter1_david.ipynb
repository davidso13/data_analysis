## 1-1. Introduction
- 첫 머신러닝 애플리케이션: 1990년대 스팸 필터(spam filter)
- 머신러닝의 정의 : 어떤 작업 T에 대한 컴퓨터 프로그램의 성능이 P일 때, 경험 E를 통해 성능이 향상되었다면 -> 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험 E로 학습한 것이다.(톰 미첼 Tom Mitchell, 1997)
- 스팸 필터에서 <br>
&nbsp;&nbsp;&nbsp;&nbsp; T: 새로운 메일이 스팸 메일인지 구분 <br>
&nbsp;&nbsp;&nbsp;&nbsp; E: 훈련 데이터 training data <br>
&nbsp;&nbsp;&nbsp;&nbsp; P: 직접 측정(정확도, accuracy) <br>
- 데이터 마이닝(data mining): 머신러닝 기술을 적용하면, 대용량 데이터에서 겉으로 보이지 않던 패턴을 발견!

### 머신러닝이 뛰어난 분야
1. 기존 솔루션으로는 많은 수동 조정과 규칙이 필요한 문제
2. 전통적인 방식으로는 해결 방법이 없는 복잡한 문제
3. 유동적인 환경(새로운 데이터에 적응)
4. 복잡한 문제, 대량의 데이터에서 통찰 얻기

### 1-2.2 배치 학습/온라인 학습 : 입력 데이터의 스트림stream부터 점진적 학습 가능 여부
(1) 배치 학습 batch learning (오프라인 학습) <br>
- 점진적으로 학습 불가능
- 시간과 자원을 많이 소모한다. -> 오프라인에서 실행
- 새로운 데이터에 대해 학습하려면 전체 데이터를 사용하여 처음부터 다시 훈련.
- 많은 컴퓨팅 자원이 필요함.(CPU, 메모리 공간, 디스크 공간, 디스크 IO, 네트워크 IO 등)
<br><br>

(2) 온라인 학습(online learning)
- 데이터를 순차적으로 한 개씩 or 미니배치라 부르는 작은 묶음 단위로 주입하여 훈련.
- 연속적으로 데이터 받고 변화에 스스로 적응해야 하는 시스템에 적합.
- 학습률(learning rate): 변화하는 데이터에 얼마나 빠르게 적응할 것인지를 나타냄.
- 학습률 증가 -> 데이터에 빠르게 적응 but, 데이터 금방 잃어버림.
- 학습률 감소 -> 느리게 학습 but, 잡음이나 대표성 없는 데이터에 덜 민감.
- 문제점: 나쁜 데이터가 주입 -> 시스템 성능이 점진적으로 감소.

## 1.3 머신러닝의 주요 도전 과제
- 나쁜 데이터
- 나쁜 알고리즘

### 1.3.1 나쁜 데이터
1. 충분하지 않은 양의 훈련 데이터
2. 대표성 없는 훈련 데이터 -> 일반화가 어려움.+ Sampling noise(우연에 의한 대표성 없는 데이터) + Sampling bias(대표성 없음, 큰 샘플)
3. 낮은 품질의 데이터 <br>
&nbsp;&nbsp;&nbsp;&nbsp; - 일부 샘플이 이상치라는 게 명확 -> 무시하거나, 수동으로 잘못된 것을  <br>
&nbsp;&nbsp;&nbsp;&nbsp; - 일부 샘플에 특성이 빠져있다면, 특성을 무시할지, 샘플을 무시할지, 빠진 값을 채울지 결정
4. 관련 없는 특성

### 1.3.2 나쁜 알고리즘
#### 1. Overfitting : test set에는 good, but 일반성 떨어짐.
- 과대적합은 (훈련 데이터의 잡음량) << (모델의 복잡성)일 떄 발생.
- 해결방법 <br>
(1) 파라미터가 적은 모델 선택 <br>
(2) 특성 수 감소 <br>
(3) 훈련 데이터 증가 <br>
(4) 훈련 데이터 잡음 감소(오류 데이터 수정, 이상치 제거) <br><br>
- 규제(Regularization): 과대적합 방지, 모델 단순 위해 모델에 제약을 가하는 것.
- 규제의 양: Hyperparmater(알고리즘의 parameter)가 결정.
<br><br>

#### 2. Underfitting : 모델이 너무 단순 -> 데이터 내재 구조 학습 불가.
- 해결방법 <br>
(1) 모델 파라미터 많은 강력한 모델 선택. <br>
(2) 알고리즘에 더 좋은 특성 제공. <br>
(3) 모델의 제약 감소(규제 hyperparameter 감소) <br>

## 1.4 테스트와 검증
- Training set, Test set으로 나누어, Generalization error 평가.
- 보통 80%/20% 비율로 사용.

### 1.4.1 Hyperparmater 튜닝과 모델 선택
1. Holdout Validation (홀드아웃 검증)
- Train set 일부를 떼어내 여러 후보 모델을 평가하고 가장 좋은 하나 선택.
- 홀드 아웃 세트 = validation set
- Train set(전체 훈련 set - validation set)에서 모델 훈련 --> validation set에서 가장 높은 성능 내는 모델 선택 --> 최선의 모델을 전체 Train set에서 다시 훈련하여 최종 모델 생성 --> 최종 모델을 Test set에서 평가.
- Validation set의 성능을 높여주기 위해 Cross-Validation 수행.(작은 검증 세트 여러 개 사용.)
