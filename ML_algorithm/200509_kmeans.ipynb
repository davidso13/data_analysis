## 1. 개요

- 대표적인 분리형 군집화 알고리즘
- 군집 수 k에 따라, 각 군집은 하나의 중심(centroid)을 가진다.
- 각 개체는 가장 가까운 중심에 할당되며, 같은 중심에 할당된 개체들이 모여 하나의 군집을 형성한다.

## 2. 학습과정

- EM 알고리즘 기반으로 작동.
- (1) 각 군집 중심의 위치 - M단계
- (2) 각 개체가 어떤 군집에 속해야 하는지 - E단계
